```python
import pandas as pd

dataset=pd.read_csv(r'C:\Users\Sandalika\Documents\machine-learning\weight-vs-height.csv').values
print(dataset,len(dataset))
```

```python
data =dataset[:,0]
target=dataset[:,1]
print('data:',data)
print('target:',target)
```

```python
#linear regression expects 2D features for the data(target can be 1D)
data=dataset[:,0].reshape(-1,1)
target=dataset[:,1]
print(data,target)
```

```python
from matplotlib import pyplot as plt
plt.scatter(data, target)
plt.xlabel('feature(weight(kg)-x)')
plt.ylabel('feature(heights(cm)-y)')
plt.plot()
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_target, test_target=train_test_split(data,target,test_size=0.1)
```

```python
#applying the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(train_data, train_target)
```

```python
# get m ad c of y=mx+c
print('m:',model.coef_)
print('c:',model.intercept_)
```

```python
# to get the m value out of the list
model.coef_[0]
```

```python
#print the values of m and c
m=model.coef_[0]
c=model.intercept_
print(m,c)
```

```python
# we get some sample x values and calculate corresponding y values
import numpy as np
xvals=np.arange(40,100)
length=len(xvals)
print('xvals:',xvals)
yvals=np.empty(length)
for i in range(length):
    yvals[i]=m*xvals[i]+c
print('-----------------------------')
print('yvals:',yvals)
    
```

```python
#draw te best fit line
plt.scatter(data,target,label='data point')
plt.plot(xvals,yvals,'r',label='Best fit line')
plt.xlabel('Feature (weights (kg)-(x)')
plt.ylabel('Label (heights (cm)-y')
plt.legend()
plt.show()
```

```python
#now we can get the predictions
test_weight=[[85]]
predicted_height=model.predict(test_weight)
print('predicted height :', predicted_height)
```
